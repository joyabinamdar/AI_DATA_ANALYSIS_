```python
# Common Data Errors Examples

# 1. Missing Data:
# Task 1: Review a dataset where some customer emails are missing. Identify how
# many records are incomplete.
# Task 2: Examine a sales dataset with missing transaction dates and determine the
# percentage of missing data.
# Task 3: Identify missing department information in an employee registry.

import pandas as pd

def identify_missing_departments(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Count the number of missing department entries
    missing_department_count = df['department'].isna().sum()
    
    # Calculate the percentage of missing departments
    total_records = len(df)
    missing_department_percentage = (missing_department_count / total_records) * 100
    
    print(f"Total records: {total_records}")
    print(f"Missing departments: {missing_department_count}")
    print(f"Percentage of missing departments: {missing_department_percentage:.2f}%")






# 2. Duplicate Data:
# Task 1: Analyze a customer dataset with duplicate entries and count the number of
# duplicates.
# Task 2: Review supplier data and identify any repeated supplier names.
# Task 3: Examine a product inventory list for duplicates in product IDs.
import pandas as pd

# Task 1: Analyze a customer dataset with duplicate entries
def analyze_customer_duplicates(file_path):
    df = pd.read_csv(file_path)
    duplicate_rows = df[df.duplicated()]
    print("Task 1: Duplicate Customer Records")
    print(f"Total duplicate rows: {len(duplicate_rows)}")
    print(duplicate_rows, end="\n\n")

# Task 2: Review supplier data and identify repeated supplier names
import pandas as pd

# Create sample customer_data.csv
customer_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 1],
    'name': ['Alice', 'Bob', 'Charlie', 'Alice'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 'alice@email.com']
})
customer_data.to_csv('customer_data.csv', index=False)

# Create sample supplier_data.csv
supplier_data = pd.DataFrame({
    'supplier_name': ['Supplier A', 'Supplier B', 'Supplier A', 'Supplier C'],
    'product': ['Product 1', 'Product 2', 'Product 3', 'Product 4'],
    'price': [20, 25, 30, 15]
})
supplier_data.to_csv('supplier_data.csv', index=False)

# Create sample product_inventory.csv
product_inventory = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P001', 'P003'],
    'name': ['Keyboard', 'Mouse', 'Keyboard', 'Monitor'],
    'stock': [50, 30, 50, 20]
})
product_inventory.to_csv('product_inventory.csv', index=False)








# 3. Inconsistent Formatting:
# Task 1: Spot inconsistencies in date formats (e.g., DD/MM/YYYY vs. MM/DD/YYYY)
# in a dataset.
# Task 2: Identify phone numbers with varying formats in a contact list.
# Task 3: Review address data for discrepancies in state abbreviations (e.g., CA vs.
# Calif.).
import pandas as pd
import re

# Task 1: Spot inconsistencies in date formats
def check_date_format_inconsistencies(df, column):
    inconsistent_dates = df[~df[column].str.match(r'^\d{2}/\d{2}/\d{4}$', na=False)]
    print("Task 1: Inconsistent Date Formats")
    print(inconsistent_dates[[column]], end="\n\n")

# Task 2: Identify inconsistent phone number formats
def check_phone_format_inconsistencies(df, column):
    # Expected format: (123) 456-7890 or 123-456-7890
    pattern = r'^(\(\d{3}\)\s\d{3}-\d{4}|\d{3}-\d{3}-\d{4})$'
    inconsistent_phones = df[~df[column].str.match(pattern, na=False)]
    print("Task 2: Inconsistent Phone Number Formats")
    print(inconsistent_phones[[column]], end="\n\n")

# Task 3: Check state abbreviation inconsistencies
def check_state_format_inconsistencies(df, column):
    valid_states = {'CA', 'NY', 'TX', 'FL', 'IL'}  # extend as needed
    inconsistent_states = df[~df[column].isin(valid_states)]
    print("Task 3: Inconsistent State Abbreviations")
    print(inconsistent_states[[column]], end="\n\n")

# Create a sample dataset
data = {
    'date_of_birth': ['12/31/1990', '31/12/1990', '1990/12/31', '12-31-1990'],
    'phone': ['123-456-7890', '(123) 456-7890', '1234567890', '123.456.7890'],
    'state': ['CA', 'Calif.', 'NY', 'Texas']
}

df = pd.DataFrame(data)

# Run inconsistency checks
check_date_format_inconsistencies(df, 'date_of_birth')
check_phone_format_inconsistencies(df, 'phone')
check_state_format_inconsistencies(df, 'state')






# 4. Data Drift:
# Task 1: Compare monthly revenues over six months to identify data drift.
# Task 2: Analyze user engagement metrics from a web application over different
# quarters.
# Task 3: Review a stock price dataset to detect any anomalies over a year.


import pandas as pd
import matplotlib.pyplot as plt

# Task 1: Compare monthly revenues
def analyze_monthly_revenue_trend(df):
    print("Task 1: Monthly Revenue Drift")
    print(df, end="\n\n")
    df.plot(x='Month', y='Revenue', kind='bar', title='Monthly Revenue')
    plt.ylabel('Revenue ($)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Task 2: Analyze user engagement metrics across quarters
def analyze_user_engagement(df):
    print("Task 2: Quarterly User Engagement")
    print(df, end="\n\n")
    df.plot(x='Quarter', y='Avg_Session_Duration', kind='line', marker='o', title='Quarterly User Engagement')
    plt.ylabel('Avg. Session Duration (minutes)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Task 3: Detect anomalies in stock prices
def detect_stock_price_anomalies(df):
    print("Task 3: Stock Price Anomaly Detection")
    print(df, end="\n\n")
    df['Anomaly'] = df['Close'].apply(lambda x: 'Anomaly' if x > df['Close'].mean() + 2*df['Close'].std() or x < df['Close'].mean() - 2*df['Close'].std() else '')
    df.plot(x='Month', y='Close', marker='o', title='Stock Price Over 12 Months')
    plt.ylabel('Stock Price ($)')
    for i, row in df.iterrows():
        if row['Anomaly']:
            plt.annotate('Anomaly', (i, row['Close']), textcoords="offset points", xytext=(0,10), ha='center', color='red')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print("Anomalies:\n", df[df['Anomaly'] == 'Anomaly'], end="\n\n")

# --- Sample Data ---

revenue_df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [10000, 10200, 9800, 12000, 15000, 11000]
})

engagement_df = pd.DataFrame({
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Avg_Session_Duration': [5.2, 4.8, 6.1, 3.9]
})

stock_df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Close': [150, 152, 149, 160, 170, 140, 138, 180, 182, 175, 200, 160]
})

# --- Execute Tasks ---
analyze_monthly_revenue_trend(revenue_df)
analyze_user_engagement(engagement_df)
detect_stock_price_anomalies(stock_df)

```
