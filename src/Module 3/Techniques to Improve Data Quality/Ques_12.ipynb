```python
# Activity 2: Dealing with Duplicates & Redundancy

# Task A: Identifying Duplicate Records

# 7. Identify Complete Duplicates:
# - Load a dataset and identify duplicated rows.
# - Use Pandas to detect duplicates.
import pandas as pd

# Sample DataFrame to simulate loading a CSV
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice'],
    'Age': [25, 30, 35, 25],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com']
}

df = pd.DataFrame(data)

# Identifying complete duplicates
duplicates = df[df.duplicated()]

print("Duplicate rows found:\n", duplicates)
print(f"Total duplicate rows: {len(duplicates)}")








# 8. Identify Duplicates based on Specific Columns:
# - Check for duplicates in specified columns.
import pandas as pd

# Sample DataFrame (use your own dataset here)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': [25, 30, 35, 25, 40, 30],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', 'eve@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

# Check for duplicates based on specific columns (e.g., 'Name' and 'Email')
duplicates = df[df.duplicated(subset=['Name', 'Email'], keep=False)]  # keep=False marks all duplicates as True

# Show the duplicate rows based on the subset of columns
print("Duplicate rows based on 'Name' and 'Email':")
print(duplicates)

# Count total duplicates based on these columns
print(f"\nTotal duplicate rows based on 'Name' and 'Email': {len(duplicates)}")







# 9. Count Duplicate Rows:
# - Calculate and print the number of duplicate rows.
import pandas as pd

# Sample DataFrame (you can replace it with your actual dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': [25, 30, 35, 25, 40, 30],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', 'eve@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

# Identify duplicate rows
duplicates = df.duplicated()

# Count the number of duplicate rows
num_duplicates = duplicates.sum()

print(f"Total number of duplicate rows: {num_duplicates}")





```

```python
# Task B: Deduplication Techniques

# 10. Remove Complete Duplicates:
# - Drop duplicate rows and keep only the first occurrence.

import pandas as pd

# Sample DataFrame (you can replace it with your actual dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': [25, 30, 35, 25, 40, 30],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', 'eve@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

# Remove complete duplicates and keep only the first occurrence
df_no_duplicates = df.drop_duplicates(keep='first')

# Show the cleaned DataFrame
print("Data after removing complete duplicates (keeping the first occurrence):")
print(df_no_duplicates)





# 11. Subset Deduplication:
# - Remove duplicates based on a subset of columns.
import pandas as pd

# Sample DataFrame (you can replace it with your actual dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': [25, 30, 35, 25, 40, 30],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', 'eve@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

# Remove duplicates based on the subset of columns (e.g., 'Name' and 'Email')
df_no_duplicates = df.drop_duplicates(subset=['Name', 'Email'], keep='first')

# Show the DataFrame after removing duplicates
print("Data after removing duplicates based on 'Name' and 'Email':")
print(df_no_duplicates)






# 12. Keep Last Occurrence:
# - Drop duplicates but keep the last occurrence in the dataset.

import pandas as pd

# Sample DataFrame (you can replace it with your actual dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Age': [25, 30, 35, 25, 40, 30],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'alice@example.com', 'eve@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

# Remove duplicates but keep the last occurrence
df_no_duplicates_last = df.drop_duplicates(keep='last')

# Show the DataFrame after keeping the last occurrence of duplicates
print("Data after removing duplicates (keeping the last occurrence):")
print(df_no_duplicates_last)




```
