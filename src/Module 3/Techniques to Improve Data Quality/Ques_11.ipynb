```python
# Activity 1: Handling Missing Data

# Task A: Dropping vs Imputation

# 1. Dropping Missing Data:
# - Load a dataset (e.g., a CSV file with some missing values like employees.csv ).
# - Inspect the dataset for missing values using a Python library (e.g., Pandas).
# - Drop rows with missing data and save the result.

import pandas as pd

# 1. Load the dataset
# Replace 'employees.csv' with the actual path to your CSV file
dataset = pd.read_csv('employees.csv')

# 2. Inspect the dataset for missing values
print("Dataset before handling missing data:")
print(dataset.isnull().sum())  # This will show the number of missing values per column

# 3. Drop rows with missing data
dataset_dropped = dataset.dropna()

# 4. Save the result to a new CSV file
dataset_dropped.to_csv('employees_dropped.csv', index=False)

print("\nDataset after dropping missing data:")
print(dataset_dropped)





# 2. Imputation using Mean:
# - Use the same dataset.
# - Fill missing numerical values with the column mean.
# - Save and display the modified data.

import pandas as pd

# 1. Load the dataset
# Replace 'employees.csv' with the actual path to your CSV file
dataset = pd.read_csv('employees.csv')

# 2. Inspect the dataset for missing values
print("Dataset before handling missing data:")
print(dataset.isnull().sum())  # This will show the number of missing values per column

# 3. Impute missing numerical values with the column mean
dataset_imputed = dataset.copy()  # To avoid modifying the original dataset

# Loop through columns and apply mean imputation for numerical columns
for column in dataset_imputed.select_dtypes(include=['float64', 'int64']).columns:
    mean_value = dataset_imputed[column].mean()
    dataset_imputed[column].fillna(mean_value, inplace=True)

# 4. Save the result to a new CSV file
dataset_imputed.to_csv('employees_imputed.csv', index=False)

# 5. Display the modified data
print("\nDataset after mean imputation:")
print(dataset_imputed)








import pandas as pd

# 1. Load the dataset
# Replace 'employees.csv' with the actual path to your CSV file
dataset = pd.read_csv('employees.csv')

# 2. Inspect the dataset for missing values
print("Dataset before handling missing data:")
print(dataset.isnull().sum())  # Show missing values for each column

# 3. Impute missing values
dataset_imputed = dataset.copy()  # To avoid modifying the original dataset

# Impute missing numerical columns with the median
for column in dataset_imputed.select_dtypes(include=['float64', 'int64']).columns:
    median_value = dataset_imputed[column].median()
    dataset_imputed[column].fillna(median_value, inplace=True)

# Impute missing categorical columns with the mode
for column in dataset_imputed.select_dtypes(include=['object']).columns:
    mode_value = dataset_imputed[column].mode()[0]
    dataset_imputed[column].fillna(mode_value, inplace=True)

# 4. Display the updated dataset
print("\nDataset after imputation (Median for numerical, Mode for categorical):")
print(dataset_imputed)




```
