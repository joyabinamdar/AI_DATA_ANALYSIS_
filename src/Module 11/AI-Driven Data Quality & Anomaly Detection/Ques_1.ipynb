# Automated Anomaly Detection
**Objective**: Understand and practice automated anomaly detection using various techniques.

**Task**: Automated Anomaly Detection using Z-score method

**Steps**:
1. Data Set: Download a dataset representing daily sales figures for a retail store.
2. Calculate Z-score: Compute the mean and standard deviation of the sales. Use these to
calculate the Z-score for each day's sales figure.
3. Identify Anomalies: Detect anomalies by identifying values with a Z-score above 3 or below -3.
4. Visualize: Plot a graph to visualize anomalies.

```python
# write your code from here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the dataset (sample data created for demo)
# You can replace this with actual CSV read, e.g., pd.read_csv('sales.csv')
dates = pd.date_range(start='2024-01-01', periods=100)
sales = np.random.normal(loc=200, scale=20, size=100)
# Inject some anomalies
sales[[10, 25, 70]] = [300, 50, 320]
df = pd.DataFrame({'date': dates, 'sales': sales})

# Step 2: Calculate Z-score
mean_sales = df['sales'].mean()
std_sales = df['sales'].std()
df['z_score'] = (df['sales'] - mean_sales) / std_sales

# Step 3: Identify Anomalies (Z-score > 3 or < -3)
df['anomaly'] = df['z_score'].apply(lambda x: 1 if abs(x) > 3 else 0)

# Step 4: Visualize
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'], label='Sales', color='blue')
plt.scatter(df[df['anomaly'] == 1]['date'], 
            df[df['anomaly'] == 1]['sales'], 
            color='red', label='Anomalies', zorder=5)
plt.title("Sales Anomaly Detection Using Z-score")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range(start='2024-01-01', periods=100)
sales = np.random.normal(loc=200, scale=20, size=100)
# Inject some anomalies
sales[[10, 25, 70]] = [300, 50, 320]
df = pd.DataFrame({'date': dates, 'sales': sales})

# Step 2: Calculate Z-score
mean_sales = df['sales'].mean()
std_sales = df['sales'].std()
df['z_score'] = (df['sales'] - mean_sales) / std_sales

# Step 3: Identify Anomalies (Z-score > 3 or < -3)
df['anomaly'] = df['z_score'].apply(lambda x: 1 if abs(x) > 3 else 0)

# Step 4: Visualize
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'], label='Sales', color='blue')
plt.scatter(df[df['anomaly'] == 1]['date'], 
            df[df['anomaly'] == 1]['sales'], 
            color='red', label='Anomalies', zorder=5)
plt.title("Sales Anomaly Detection Using Z-score")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
