```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
import matplotlib.pyplot as plt
import numpy as np

# Data for product sales (Bar Plot)
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [250, 400, 150, 300]

# Data for height vs. weight (Scatter Plot)
heights = [5.5, 6.0, 5.8, 5.9, 5.7]
weights = [150, 180, 160, 175, 165]

# Create a figure and a set of subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Bar Plot for Product Sales
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales')

# Scatter Plot for Height vs. Weight
axs[1].scatter(heights, weights, color='red')
axs[1].set_title('Height vs. Weight')
axs[1].set_xlabel('Height (ft)')
axs[1].set_ylabel('Weight (lbs)')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()

```
