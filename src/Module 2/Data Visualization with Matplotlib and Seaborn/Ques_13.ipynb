```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a customized boxplot for petal length by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris, 
            palette='Set2',  # Set a color palette
            width=0.5,       # Box width
            fliersize=7,     # Size of outliers
            linewidth=2,     # Width of the lines
            notch=True)      # Add notches to the box

# Customize labels and title
plt.title('Distribution of Petal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=14)
plt.ylabel('Petal Length (cm)', fontsize=14)

# Show the plot
plt.show()
```
