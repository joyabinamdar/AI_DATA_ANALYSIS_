```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Customize style
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='darkcyan', edgecolor='black')

# Add labels and title with custom fonts
plt.title('Customized Product Sales (in Thousands)', fontsize=16, fontweight='bold', fontname='Arial')
plt.xlabel('Product', fontsize=13, fontstyle='italic', fontname='Arial')
plt.ylabel('Sales', fontsize=13, fontstyle='italic', fontname='Arial')

# Grid
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

# Add value labels on top of bars
for i, value in enumerate(sales):
    plt.text(i, value + 2, str(value), ha='center', fontsize=11)

# Show plot
plt.tight_layout()
plt.show()
```
