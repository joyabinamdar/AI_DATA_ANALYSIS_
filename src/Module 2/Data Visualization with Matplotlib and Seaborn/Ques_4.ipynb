```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create DataFrame
data = pd.DataFrame({
    'Product': products,
    'Sales': sales
})

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x='Product', y='Sales', palette='Blues_d')

# Add titles and labels
plt.title('Product Sales (in Thousands)', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```
