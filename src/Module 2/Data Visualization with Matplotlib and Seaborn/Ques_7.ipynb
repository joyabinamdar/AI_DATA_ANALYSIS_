```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]  # Size of points

# Create DataFrame
data = pd.DataFrame({
    'Height (cm)': heights,
    'Weight (kg)': weights,
    'Size': sizes
})

# Create advanced scatter plot with point size
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Height (cm)', y='Weight (kg)', size='Size', sizes=(50, 200), color='blue', legend=None)

# Add titles and labels
plt.title('Advanced Height vs Weight Scatter Plot (Point Size Based on Size)', fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```
