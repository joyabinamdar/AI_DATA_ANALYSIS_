```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create DataFrame
data = pd.DataFrame({
    'Height (cm)': heights,
    'Weight (kg)': weights
})

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Height (cm)', y='Weight (kg)', color='teal', edgecolor='black')

# Add titles and labels
plt.title('Height vs Weight Scatter Plot (Seaborn)', fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```
