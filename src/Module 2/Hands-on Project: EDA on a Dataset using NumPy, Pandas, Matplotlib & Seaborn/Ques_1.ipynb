```python
# Title: Load & Examine a Basic Dataset
# Description: Load an example dataset such as the Iris dataset and perform basic exploration.
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Series(iris.target).map({i: name for i, name in enumerate(iris.target_names)})

# Display first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Dataset shape
print("\nShape of the dataset:")
print(df.shape)

# Data types and non-null values
print("\nDataset info:")
print(df.info())

# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```
