```python
# Question: Visualizing Outliers Using Boxplots
# Description: Use a boxplot to visually detect outliers in a dataset.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate a sample dataset
np.random.seed(42)
data = {
    'Salary': np.append(np.random.normal(50000, 5000, 100), [20000, 120000])  # Add two outliers
}
df = pd.DataFrame(data)

# Step 2: Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Salary'], color='skyblue')
plt.title('Boxplot to Detect Outliers in Salary')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```
