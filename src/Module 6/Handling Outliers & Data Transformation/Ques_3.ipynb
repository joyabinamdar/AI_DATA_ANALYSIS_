```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with positive values
data = {
    'Salary': [25000, 27000, 29000, 31000, 33000, 35000, 40000, 45000, 100000]
}
df = pd.DataFrame(data)

# Check original distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Salary'], kde=True, bins=10)
plt.title("Original Salary Distribution")

# Apply Box-Cox transformation (data must be positive)
salary_transformed, lambda_val = stats.boxcox(df['Salary'])

# Add transformed data to DataFrame
df['Salary_BoxCox'] = salary_transformed

# Check transformed distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Salary_BoxCox'], kde=True, bins=10, color='green')
plt.title("Box-Cox Transformed Salary")

plt.tight_layout()
plt.show()

# Print lambda value used for transformation
print(f"Optimal  (lambda) for Box-Cox transformation: {lambda_val:.4f}")

```
