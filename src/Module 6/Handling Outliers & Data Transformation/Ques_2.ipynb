```python
# Question: Winsorization of Data
# Description: Apply Winsorization to a given dataset to handle outliers.

import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt

# Sample dataset with outliers
data = {'Salary': [35000, 42000, 41000, 39000, 50000, 47000, 900000, 43000, 41000, 32000, 870000]}
df = pd.DataFrame(data)

# Apply Winsorization to limit the top and bottom 5% of the data
# limits=(lower, upper) as proportions of data
df['Winsorized_Salary'] = winsorize(df['Salary'], limits=[0.05, 0.05])

# Display before and after
print("Original Salaries:\n", df['Salary'])
print("\nWinsorized Salaries:\n", df['Winsorized_Salary'])

# Optional: Visualize the effect
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(df['Salary'])
plt.title("Original Salary Data")

plt.subplot(1, 2, 2)
plt.boxplot(df['Winsorized_Salary'])
plt.title("Winsorized Salary Data")

plt.tight_layout()
plt.show()
```
